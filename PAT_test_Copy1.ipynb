> Solving for Q 1 :Calculate average Stock Grant by job level, location and fiscal quarter

```python
# Cleaning & making data compatible to software used  is a crucial Task 
# Since the Column headings have Spaces and this can cause issue during indexing of columns , 
#    we can rename using the below lambda function



cols = PAT_df.columns
cols = cols.map(lambda x: x.replace(' ', '_') if isinstance(x, (str, unicode)) else x)
PAT_df.columns = cols
```

```python
cols_to_keep1 = ['Quarter','Job_Level','Location','Stock_Amount'] 
stock_amt = PAT_df[cols_to_keep1]
```

```python
level_df = pd.DataFrame(stock_amt)
```

```python
level_df.head()
```

```python
 yu = level_df.groupby(['Quarter'], as_index=1).mean()
```

```python
print(yu)
```

```python
tal_df= pd.DataFrame(yu)
```

```python
jel = tal_df.sort_values(by=['Stock_Amount'], axis=0, ascending=True)
```

```python
my_plot = jel.plot(kind='bar',figsize=(10,8),color ="orange", title ="Stock Grant for  2015-2016")
my_plot.legend(["Stock_Amount"], loc='upper left')
```

Q1 of 2016 had the lowest Stock grant amount among all the other Quarters an Average of USD 2282 , while Q2 of 2016 
had the highest Stock grant of USD 3369

```python
yu2 = level_df.groupby(['Quarter','Job_Level'], as_index=1).mean()
```

```python
yu2.head()
```

```python
yu2.unstack().head()
```

```python
colors88 = plt.cm.gist_rainbow(np.linspace(0, 1, 4))
```

```python
my_plot = yu2.unstack().plot(kind='bar',stacked=True,title="Avg Stock Grant by Qrts for Job Levels ", figsize=(10,8), color =colors88)
my_plot.set_xlabel("Quarters")
my_plot.set_ylabel("Job_Level")
```

Job Level-1 did not get any Stock Grant for Q4 2015 , while in all previous Quarters Job level -1 has been getting 
USD 922 on average.  While Job Level -4 has been getting on average 6914 for all quarters
